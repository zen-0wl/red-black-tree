```python
import sys

#node creation
class Node():
    def __init__(self, type):
        self.type = type
        self.colour = 1
        self.parent = None
        self.left = None
        self.right = None 
#class for red black tree 
class RBT():
    def __init__(self):
        self.Tempty = Node(0)
        self.Tempty.left = None
        self.Tempty.right = None
        self.Tempty.colour = 0
        self.root = self.Tempty

    # preorder
    def pre_order(self, node):
        if node != Tempty:
            print(node.item + " ")
            self.pre_order(node.left)
            self.pre_order(node.right)
    # inorder 
    def in_order(self, node):
        if node != Tempty:
            self.in_order(node.left)
            print(node.item + " ")
            self.in_order(node.right)
    # postorder
    def post_order(self, node):
        if node != Tempty:
            self.post_order(node.left)
            self.post_order(node.right)
            print(node.item + " ")
    
    # search operation of red black tree
    def searchTree(self, node, key):
        if node == Tempty or key == node.type:
            return node 
    
    # balance insert operation of red black tree 
    def insertion(self, i):
        while i.parent.colour == 1:
            if i.parent == i.parent.parent.left:
                x = i.parent.parent.left

                if x.colour == 1:
                    x.colour == 0
                    i.parent.colour = 0
                    i.parent.parent.colour = 1
                    i = i.parent.parent 
                elif i == i.parent.left:
                    i = i.parent
                    self.rotateRight(i)
                i.parent.colour = 0
                i.parent.parent.colour = 1
                self.rotateLeft(i.parent.parent)
            else:
                x = i.parent.parent.right 

                if x.colour == 1:
                    x.colour = 0
                    i.parent.colour = 0
                    i.parent.parent.colour = 1
                    i = i.parent.parent
                elif i == i.parent.right:
                    i = i.parent
                    self.rotateLeft(i)
                i.parent.colour = 0
                i.parent.parent.colour = 1
                self.rotateRight(i.parent.parent)
            # if statement for root balance
            if i == self.root:
                break
        self.root.colour = 0

    # Tree output 
    def printTree(self, node, indent, last):
        if node != self.Tempty:
            sys.stdout.write(indent)
            if last:
                sys.stdout.write("R----")
                indent += "     "
            else:
                sys.stdout.write("L----")
                indent += "|    "
            
            v_colour = "RED" if node.colour == 1 else "BLACK"
            print(str(node.type) + "(" + v_colour + ")")
            self.printTree(node.left, indent, False)
            self.printTree(node.right, indent, True)

    def preorder(self):
        self.pre_order(self.root)
    
    def inorder(self):
        self.in_order(self.root)

    def postorder(self):
        self.post_order(self.root)

    def search(self, i):
        return self.searchTree(self.root, i)

    def min(self, node):
        while node.left != self.Tempty:
            node = node.left
        return node

    def max(self, node):
        while node.right != self.Tempty:
            node = node.right
        return node

    # left rotation
    def rotateLeft(self, x):
        y = x.right
        x.right = y.left
        if y.left != self.Tempty:
            y.left.parent = x

        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y
        y.left = x
        x.parent = y

    # right rotation
    def rotateRight(self, x):
        y = x.left
        x.left = y.right
        if y.right != self.Tempty:
            y.right.parent = x

        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.right:
            x.parent.right = y
        else:
            x.parent.left = y
        y.right = x
        x.parent = y
    
    # Insertion of Red Black Tree
    def insert(self, key):
        node = Node(key)
        node.parent = None
        node.type = key
        node.left = self.Tempty
        node.right = self.Tempty
        node.colour = 1

        y = None
        x = self.root

        while x != self.Tempty:
            y = x
            if node.type < x.type:
                x = x.left
            else:
                x = x.right

        node.parent = y
        if y == None:
            self.root = node
        elif node.type < y.type:
            y.left = node
        else:
            y.right = node

        if node.parent == None:
            node.colour = 0
            return

        if node.parent.parent == None:
            return

        self.insertion(node)

    def get_root(self):
        return self.root

    def print_tree(self):
        self.printTree(self.root, "", True)

# Executes main function
if __name__ == "__main__":
    bst = RBT()

    bst.insert(2)
    bst.insert(13)
    bst.insert(7)
    bst.insert(16)
    bst.insert(19)
    bst.insert(9)
    bst.insert(22)
    bst.insert(10)
    bst.insert(14)
    bst.insert(17)

    bst.print_tree()
```
